# Linear Regression

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Mall_Customers.csv')
data
```

```python
x = data[['Annual Income (k$)']].values
y = data['Spending Score (1-100)'].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
import matplotlib.pyplot as pyplot
pyplot.xlabel("Annual Income")
pyplot.ylabel("Spending Score")
pyplot.scatter(x,y)
pyplot.plot(x,lin_reg.predict(x),color='black')
pyplot.title('Linear Regression Line')
pyplot.show()
```

```python
m = lin_reg.coef_
```

```python
c = lin_reg.intercept_
```

```python
y = m*x+c
y
```
